# Illustrative Yield Curves
### Plot a Normal Yield Curve

```python
## Plot a normal yield curve

import matplotlib.pyplot as plt

# Set the style with Seaborn
import seaborn as sns
sns.set(style='dark', palette='deep', font='sans-serif')

# Data for the chart
bond_maturities = [1, 2, 3, 5, 7, 10, 20, 30]
bond_yields = [1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5]

# Create the chart
plt.plot(bond_maturities, bond_yields)

# Add labels to the axes
plt.xlabel("Bond Maturity (years)")
plt.ylabel("Bond Yield (%)")

# Show the plot
plt.show()
```

### Plot Normal, Inverted, & Flat Yield Curves

```python
## Plot a normal, inverted, and flat yield curve

import pandas as pd
import matplotlib.pyplot as plt


# Set up the data for the normal yield curve
normal_data = {
    "Maturity": [1, 2, 3, 5, 7, 10, 20, 30],  # maturities in years
    "Yield": [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5],  # yields in percentage points
}

# Set up the data for the inverted yield curve
inverted_data = {
    "Maturity": [1, 2, 3, 5, 7, 10, 20, 30],  # maturities in years
    "Yield": [4.5, 4, 3.5, 3, 2.5, 2, 1.5, 1],  # yields in percentage points
}

# Set up the data for the flat yield curve
flat_data = {
    "Maturity": [1, 2, 3, 5, 7, 10, 20, 30],  # maturities in years
    "Yield": [3, 3, 3, 3, 3, 3, 3, 3],  # yields in percentage points
}

# Create the dataframes
normal_df = pd.DataFrame(normal_data)
inverted_df = pd.DataFrame(inverted_data)
flat_df = pd.DataFrame(flat_data)

# Create the figure and subplots
fig, axs = plt.subplots(1, 3, figsize=(20, 3))

# Plot the normal yield curve
normal_df.plot(x="Maturity", y="Yield", kind="line", ax=axs[0], color="cornflowerblue", linewidth=5)
axs[0].set_title("Normal Yield Curve")
axs[0].set_xlabel("Maturity (years)")
axs[0].set_ylabel("Yield (%)")

# Plot the inverted yield curve
inverted_df.plot(x="Maturity", y="Yield", kind="line", ax=axs[1], color="navy", linewidth=5)
axs[1].set_title("Inverted Yield Curve")
axs[1].set_xlabel("Maturity (years)")
axs[1].set_ylabel("Yield (%)")

# Plot the flat yield curve
flat_df.plot(x="Maturity", y="Yield", kind="line", ax=axs[2], color="steelblue", linewidth=5)
axs[2].set_title("Flat Yield Curve")
axs[2].set_xlabel("Maturity (years)")
axs[2].set_ylabel("Yield (%)")

# Save the plot as a PNG file
plt.savefig('yieldcurves_multiplot.png', dpi=300)

# Show the plots
plt.show()

```
