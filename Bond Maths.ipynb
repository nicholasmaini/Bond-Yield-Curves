# Bond Maths
### Plotting a Bond Price-Yield Relationship

```python
## Bond Price-Yield Relationship

import matplotlib.pyplot as plt

# Set the style with Seaborn
import seaborn as sns
sns.set(style='dark', palette='deep', font='sans-serif')

# Data for the chart
bond_yields = [3.2, 3.0, 2.8, 2.6, 2.4, 2.2]
bond_prices = [90, 92, 94, 96, 98, 100]

plt.figure(figsize=(10, 5))

# Create the chart
plt.plot(bond_yields, bond_prices)

# Add labels to the axes
plt.xlabel("Bond Yield (%)")
plt.ylabel("Bond Price")

# Add a title
plt.title("Bond Price-Yield Relationship")

# Show the plot
plt.show()
```

### Plotting a Bond Price-Yield Relationship over Time 

```python
## Bond Price-Yield Relationship over Time - Straight-Line


import matplotlib.pyplot as plt

# Data for the first line
time = [1, 2, 3, 4, 5]
bond_yields = [3.2, 3.0, 2.8, 2.6, 2.4]

# Calculate bond prices based on bond yields
bond_prices = [100 - y for y in bond_yields]

# Create the chart
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot bond yield on the first y-axis
ax1.plot(time, bond_yields, color='b')
ax1.set_xlabel("Time")
ax1.set_ylabel("Bond Yield", color='b')

# Add a second y-axis for bond price
ax2 = ax1.twinx()
ax2.plot(time, bond_prices, color='r')
ax2.set_ylabel("Bond Price", color='r')

# Add a title
plt.title("Bond Price-Yield Relationship over Time")

# Show the plot
plt.show()
```

```python
## Bond Price-Yield Relationship over Time - Randomised

import matplotlib.pyplot as plt
import random

# Generate 50 random bond yields
bond_yields = [random.uniform(3.0, 3.2)]
for i in range(199):
    bond_yields.append(random.uniform(bond_yields[-1] * 0.9, bond_yields[-1] * 1.1))

# Calculate bond prices based on bond yields
bond_prices = [100 - y for y in bond_yields]

# Create the chart
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot bond yield on the first y-axis
ax1.plot(range(200), bond_yields, color='b')
ax1.set_xlabel("Time")
ax1.set_ylabel("Bond Yield", color='b')

# Add a second y-axis for bond price
ax2 = ax1.twinx()
ax2.plot(range(200), bond_prices, color='r')
ax2.set_ylabel("Bond Price", color='r')

# Add a title
plt.title("Bond Price-Yield Relationship over Time")

# Show the plot
plt.show()
```

### Plotting a Bond Price-Yield Relationship with Convexity
